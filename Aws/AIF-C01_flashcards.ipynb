# AWS Certified AI Practitioner (AIF-C01) フラッシュカード
このノートブックは10問のフラッシュカードをJSONから読み込み、Q/Aを確認できる簡易クイズ用セルを提供します。

```python
import json, random
with open(r'./aif-c01_flashcards.json', 'r', encoding='utf-8') as f:
    cards = json.load(f)
print(f'読み込んだカード枚数: {len(cards)}')
random.seed(0)
random.shuffle(cards)
cards[:2]  # 最初の2枚をプレビュー
```

## 単発クイズ実行

```python
def ask(i):
    c = cards[i % len(cards)]
    print(f"Q{i+1}: {c['question_ja']}")
    for k,v in c['choices'].items():
        print(f"  {k}. {v}")
    # ここで input() を使わず、解答と解説をそのまま表示
    print('\n[正解]', c['answer'])
    print('[解説]', c['explanation_ja'])

for i in range(len(cards)):
    ask(i)
```
