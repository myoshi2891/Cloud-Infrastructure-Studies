# AWS Certified AI Practitioner (AIF-C01) フラッシュカード v2
- 合計: 110問
- 言語: 日本語/英語併記
- 難易度: 用語 → 設計 → ケース
- 機能: フィルタ、シャッフル、小テスト(採点)

```python
import json, random
with open(r'./aif-c01_flashcards_110.json', 'r', encoding='utf-8') as f:
    cards = json.load(f)
print('Loaded cards:', len(cards))
```

## フィルタ & シャッフル設定

```python
def filter_cards(domain=None, difficulty=None):
    result = cards
    if domain:
        result = [c for c in result if domain in c['domain']]
    if difficulty:
        result = [c for c in result if c['difficulty'] == difficulty]
    return result

# 例: 難易度'設計'に絞ってシャッフル
subset = filter_cards(difficulty='設計')
random.seed(0)
random.shuffle(subset)
len(subset)
```

## 単発Q/A（自動表示）

```python
def show_card(c):
    print(c['question_ja'])
    print('(EN):', c['question_en'])
    for k in ['A','B','C','D']:
        print(f"  {k}. {c['choices_ja'][k]} (EN: {c['choices_en'][k]})")
    print('[Answer]', c['answer'])
    print('[解説]', c['explanation_ja'])
    print('[Explanation]', c['explanation_en'])

for i,c in enumerate(subset[:30]):
    print(f'--- Card #{i+1} ---')
    show_card(c)
```

## 小テスト（採点付き）

```python
def quiz(num_questions=10, seed=0, domain=None, difficulty=None):
    qs = filter_cards(domain=domain, difficulty=difficulty)
    if not qs:
        print('対象カードがありません。フィルタ条件を見直してください。')
        return
    random.seed(seed)
    random.shuffle(qs)
    qs = qs[:num_questions]
    correct = 0
    answers = []
    for i,c in enumerate(qs, 1):
        print(f'Q{i}: {c["question_ja"]}')
        print('(EN):', c['question_en'])
        for k in ['A','B','C','D']:
            print(f"  {k}. {c['choices_ja'][k]} (EN: {c['choices_en'][k]})")
        user = input('あなたの解答 (A/B/C/D): ').strip().upper()
        answers.append((user, c['answer']))
        if user == c['answer']:
            print('✔ 正解!')
            correct += 1
        else:
            print(f'✘ 不正解。正解は {c["answer"]}')
        print('解説:', c['explanation_ja'])
        print('Explanation:', c['explanation_en'])
        print('-'*40)
    score = 100.0 * correct / len(qs)
    print(f'正解数: {correct}/{len(qs)} | スコア: {score:.1f}点')
    return score, answers

# 例: 10問、シード=1、難易度'用語'に絞る
# quiz(10, seed=1, difficulty='用語')
```
